```python
fn = "HeidelbergWallboxEnergyControl_ModbusRegisterTable.pdf"
doc = fitz.open(fn)
```

```python
doc.metadata
```

```python
page = doc.loadPage(0)
```

```python
blocks = page.get_text("dict")["blocks"]
len(blocks)
```

```python
for i, block in enumerate(blocks):
    if "lines" in block:
        for j, item in enumerate(block["lines"]):
            try:
                print(i, j, item["spans"][0]["text"])
            except:
                print(i, j)
    else:
        print(i)
```

```python
texts = []
for i, block in enumerate(blocks):
    if "lines" in block:
        for j, item in enumerate(block["lines"]):
            try:
                texts.append(item["spans"][0]["text"])
            except:
                pass
```

```python
texts
```

```python
lines = []
new_line = True
for text in texts:
    if new_line:
        line = []
        new_line = False
    line.append(text)
    for endline in ["V1.", "V 1.", "Available "]:
        if text.startswith(endline):
            new_line = True
            lines.append(line)
```

```python
for i, line in enumerate(lines):
    print(i, len(line))
```
